# Popularity Based Recommender System

```python
ratings_with_name = ratings.merge(books,on="ISBN")
```

```python
num_rating_df = ratings_with_name.groupby("Book-Title")["Book-Rating"].count().reset_index()
num_rating_df.rename(columns={"Book-Rating":"num_ratings"}, inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby("Book-Title")["Book-Rating"].mean().reset_index()
avg_rating_df.rename(columns={"Book-Rating":"avg_ratings"}, inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on="Book-Title")
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings'] >= 300].sort_values("avg_ratings",ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books, on="Book-Title").drop_duplicates("Book-Title")
popular_df = popular_df[["Book-Title","Book-Author","Image-URL-M","num_ratings","avg_ratings"]]
```

```python
popular_df
```

# Collaborative Filtering Based Recommender System

```python
x = ratings_with_name.groupby('User-ID')['Book-Rating'].count() > 200
imp_users = x[x].index
```

```python
filtered_ratings = ratings_with_name[ratings_with_name["User-ID"].isin(imp_users)]
```

```python
y = filtered_ratings.groupby('Book-Title')['Book-Rating'].count() >= 50
famous_books = y[y].index
```

```python
final_ratings = filtered_ratings[filtered_ratings["Book-Title"].isin(famous_books)]
```

```python
final_ratings
```

```python
final_ratings.isnull().sum()
```

```python
final_ratings.duplicated().sum()
```

```python
final_ratings["Book-Title"].nunique()
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values="Book-Rating")
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

## Model building on the bases of similarity score  

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x:x[1], reverse=True)[1:9]
    
    for i in similar_items:
        print(pt.index[i[0]])
    
```

```python
recommend("Message in a Bottle")
```
